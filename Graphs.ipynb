```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv('Indian-Export.csv')
x = data['High-technology exports (% of manufactured exports)']
y = data['Merchandise exports (current US$)']
z = data['Agricultural raw materials exports (% of merchandise exports)']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.set_xlabel('High-technology exports')
ax.set_ylabel('Merchandise exports')
ax.set_zlabel('Agricultural raw materials exports')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Indian-Export.csv')
fig = plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=2)
ax2 = plt.subplot2grid((2, 2), (1, 0))
ax3 = plt.subplot2grid((2, 2), (1, 1))
df.plot(x='YEAR', y='High-technology exports (% of manufactured exports)', ax=ax1)
ax1.set_title('High-technology exports')
df.plot(x='YEAR', y='Merchandise exports (current US$)', ax=ax2)
ax2.set_title('Merchandise exports')
df.plot(x='YEAR', y='Agricultural raw materials exports (% of merchandise exports)', ax=ax3)
ax3.set_title('Agricultural raw materials exports')
plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Indian-Export.csv")
cols = ["High-technology exports (% of manufactured exports)", "Merchandise exports (current US$)", "Agricultural raw materials exports (% of merchandise exports)"]
corr_matrix = df[cols].corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('Indian-Export.csv')
x_col = 'Merchandise exports to low- and middle-income economies in Europe & Central Asia (% of total merchandise exports)'
y_col = 'Merchandise exports to low- and middle-income economies in Sub-Saharan Africa (% of total merchandise exports)'
x_data = df[x_col]
y_data = df[y_col]
sns.regplot(x=x_data, y=y_data)
plt.title("MERCHANDISE EXPORTS IN EUROPE & CENTRAL ASIA vs MERCHANDISE EXPORTS IN SUB-SAHARAN AFRICA")
plt.xlabel("MERCHANDISE EXPORTS IN EUROPE & CENTRAL ASIA")
plt.ylabel("MERCHANDISE EXPORTS IN SUB-SAHARAN AFRICA")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import axes3d

data = pd.read_csv("Indian-Export.csv")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = data['Export value index (2000 = 100)'].values
y = data['Net barter terms of trade index (2000 = 100)'].values
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

ax.plot_wireframe(X, Y, Z, rstride=2, cstride=2)

ax.set_xlabel('Export value index (2000 = 100)')
ax.set_ylabel('Net barter terms of trade index (2000 = 100)')
ax.set_zlabel('')

plt.title("Wireframe Plot Example")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show
data = pd.read_csv('Indian-Export.csv', index_col=False)
data = data.rename(columns={
    'Insurance and financial services (% of commercial service exports)': 'Insurance and financial services',
})
output_file("insurance_financial_services.html")
p = figure(title="Insurance and financial services (% of commercial service exports)", x_axis_label='YEAR', y_axis_label='Insurance and financial services (% of commercial service exports)')
p.line(data['YEAR'], data['Insurance and financial services'], legend_label="Insurance and financial services", line_width=2)
show(p)
```
